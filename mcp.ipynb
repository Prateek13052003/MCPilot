```python
from mcp import ClientSession
from mcp.client.stdio import stdio_client, StdioServerParameters

print("MCP imports successful")
```

```python
server_params = StdioServerParameters(
    command="npx",
    args=[
        "-y",
        "@modelcontextprotocol/server-filesystem",
        "."
    ]
)

print(server_params)
```

```python
async def test_connection():
    async with stdio_client(server_params) as (read, write):
        async with ClientSession(read, write) as session:

            await session.initialize()

            print("✅ Connected to MCP Filesystem Server")

await test_connection()
```

```python
async def list_tools():
    async with stdio_client(server_params) as (read, write):
        async with ClientSession(read, write) as session:

            await session.initialize()

            tools = await session.list_tools()

            print("Available Tools:\n")

            for tool in tools.tools:
                print("-", tool.name)

await list_tools()
```

```python
async def get_allowed_directories():
    async with stdio_client(server_params) as (read, write):
        async with ClientSession(read, write) as session:

            await session.initialize()

            result = await session.call_tool(
                "list_allowed_directories",
                {}
            )

            return result

result = await get_allowed_directories()
print(result)
```

```python
import ollama
def ask_ollama(prompt):
    response = ollama.chat(
        model="qwen2.5:7b",
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ]
    )

    return response["message"]["content"]
```

```python
response = ask_ollama(
    "What files might be useful in a Python project?"
)

print(response)
```

```python
async def list_project_files():

    async with stdio_client(server_params) as (read, write):
        async with ClientSession(read, write) as session:

            await session.initialize()

            result = await session.call_tool(
                "list_directory",
                {
                    "path": "/Users/prateekchoudhary/Downloads/agents-main/6_mcp"
                }
            )

            return str(result)
```

```python
async def mcp_agent(question):

    files = await list_project_files()

    prompt = f"""
You are an AI software engineer.

A filesystem MCP server returned the following project information:

{files}

User Question:
{question}

Answer using the project information.
"""

    return ask_ollama(prompt)
```

```python
result = await mcp_agent(
    "What files are present in this project and what does this project appear to be about?"
)

print(result)
```

```python
async def read_file(path):

    async with stdio_client(server_params) as (read, write):
        async with ClientSession(read, write) as session:

            await session.initialize()

            result = await session.call_tool(
                "read_file",
                {
                    "path": path
                }
            )

            return str(result)
```

```python
content = await read_file(
    "/Users/prateekchoudhary/Downloads/agents-main/6_mcp/app.py"
)

print(content[:3000])
```

```python
async def analyze_project():

    app = await read_file(
        "/Users/prateekchoudhary/Downloads/agents-main/6_mcp/app.py"
    )

    database = await read_file(
        "/Users/prateekchoudhary/Downloads/agents-main/6_mcp/database.py"
    )

    market = await read_file(
        "/Users/prateekchoudhary/Downloads/agents-main/6_mcp/market.py"
    )

    prompt = f"""
You are a Senior Software Architect.

Analyze this codebase.

APP.PY
{app}

DATABASE.PY
{database}

MARKET.PY
{market}

Provide:

1. Project Overview
2. Architecture Diagram (Text)
3. Purpose of Each File
4. Data Flow
5. Key Components
6. Potential Improvements
7. Scalability Suggestions
"""

    return ask_ollama(prompt)
```

```python
result = await analyze_project()

print(result)
```
